```python
import import_ipynb
import random
import math 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.special import logsumexp
import utils as ut
from sklearn.model_selection import train_test_split
```

```python
data = [ut.dataset.get(False)]
df = pd.DataFrame({
    "x":data[0][0],
    "y":data[0][1],
    "label":data[0][2]
                  })
print(df.head())
```

```python
graph = sns.scatterplot(data=df,x ="x",y="y",hue="label")
graph.set_xlim(-4,4)
X = df.iloc[: , :-1]
X = np.array(X)
y = df.iloc[:,-1:]
y = np.array(y)
p = ut.Perceptron(learning_rate=0.05, epochs=1000)
p.fit(X, y)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X[:, 0], X[:, 1], marker="o", c=y)
x1 = np.amin(X[:, 0])
x2 = np.amax(X[:, 0])
x3 = (-p.wt[0] * x1 - p.bias) / p.wt[1]
x4= (-p.wt[0] * x2- p.bias) / p.wt[1]
ax.plot([x1, x2], [x3, x4], "green",linestyle = 'dashed')
ax.set_xlim([-4, 4])
plt.title("Variable bias without noise decision boundary")
plt.show()
```
